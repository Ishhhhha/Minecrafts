**STUDENT PERFORMANCE DATA ANALYSIS**

##### Step 1: Load the Dataset

We load the student performance dataset using pandas to begin our analysis.

```python
import pandas as pd
df = pd.read_csv(
    r"C:\Users\admin\Documents\ISHA\Minecraft\student-mat.csv",
    sep=";"
)
 
df.head()
```

 ##### Step 2: Explore and Clean the Data

In this step we:
- Check for missing values
- Remove duplicate records
- Inspect dataset shape and data types

```python
df.shape
df.dtypes
df.isnull().sum()
df = df.drop_duplicates()
df.shape
```

##### Step 3: Data Analysis

We answer the following questions:
1. What is the average final grade (G3)?
2. How many students scored above 15?
3. Is study time correlated with performance?
4. Which gender performs better on average?

```python
df["G3"].mean()
```

```python
(df["G3"] > 15).sum()
```

```python
df["studytime"].corr(df["G3"])
```

```python
df.groupby("sex")["G3"].mean()
```

##### Step 4: Data Visualization

We visualize:
- Distribution of final grades
- Relationship between study time and grades
- Gender-wise average final scores

```python
import matplotlib.pyplot as plt

plt.hist(df["G3"])
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

```python
plt.scatter(df["studytime"], df["G3"])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
gender_avg = df.groupby("sex")["G3"].mean()

gender_avg.plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.title("Average Final Grade by Gender")
plt.show()
```

##### Conclusion


This analysis examined student performance data to identify factors influencing final grades. After cleaning the dataset, key insights were derived from statistical analysis and visualizations. The results show that overall performance varies, with fewer students achieving high scores. Study time has a weak to moderate positive relationship with final grades, indicating it helps but is not the only determining factor. Differences in average performance between genders were observed, highlighting trends without implying causation. Visualizations supported better understanding of these patterns.
